### Importación de datos

#1. Análisis de facturación



#### En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_total1 = tienda['Precio'].sum()
ingreso_total1
```

```python
ingreso_total2 = tienda2['Precio'].sum()
ingreso_total2
```

```python
ingreso_total3 = tienda3['Precio'].sum()
ingreso_total3
```

```python
ingreso_total4 = tienda4['Precio'].sum()
ingreso_total4
```

```python
print(f'El ingreso de la tienda 1 fue de: ${ingreso_total1:,.0f} pesos')
print(f'El ingreso de la tienda 2 fue de: ${ingreso_total2:,.0f} pesos')
print(f'El ingreso de la tienda 3 fue de: ${ingreso_total3:,.0f} pesos')
print(f'El ingreso de la tienda 4 fue de: ${ingreso_total4:,.0f} pesos')
```

```python
import matplotlib.pyplot as plt
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total1, ingreso_total2, ingreso_total3, ingreso_total4]
colores = ['#FFD07F', '#FDA65D', '#FF8243', '#E26A2C']

plt.figure(figsize=(14,6))
barras = plt.bar(tiendas, ingresos, color=colores, edgecolor='black', linewidth=1.5)
plt.subplot(1,1,1)
plt.title('Ingreso total de cada tienda', fontsize=14 )
plt.ylabel('Ingresos', fontsize=12)

for barra in barras:
  altura = barra.get_height()
  plt.text(barra.get_x() + barra.get_width()/2., altura, f'${altura/1e6:.1f}M', ha='center', va='bottom', fontsize=10,)
```

Descripción: Se coloca en la gráfica los colores del más claro a la tienda que cuenta con ingresos más elevados y va subiendo de color conforme el ingreso va disminuyendo, colocando el foco al color más oscuro.  

# 2. Ventas por categoría

#### En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ven_cat_t1 = tienda.groupby('Categoría del Producto')['Categoría del Producto'].count().sort_values(ascending=False)
ven_cat_t1

new_var = plt.figure(figsize=(6,6))
# Lista de colores:
pie_colors = ['#711DB0', '#C21292', '#EF4040', '#FFA732', '#F5E965', '#FF9668', '#DB456F', '#A74FAF']
plt.pie(ven_cat_t1.values, labels=ven_cat_t1.index, autopct='%1.1f%%', shadow=True, startangle=140, colors=pie_colors)
plt.title('Ventas por categoría tienda 1')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
ven_cat_t3 = tienda3.groupby('Categoría del Producto')['Categoría del Producto'].count().sort_values(ascending=False)
ven_cat_t3


new_var = plt.figure(figsize=(6,6))
# Lista de colores:
pie_colors = ['#711DB0', '#C21292', '#EF4040', '#FFA732', '#F5E965', '#FF9668', '#DB456F', '#A74FAF']
plt.pie(ven_cat_t3.values, labels=ven_cat_t3.index, autopct='%1.1f%%', shadow=True, startangle=140, colors=pie_colors)
plt.title('Ventas por categoría tienda 3')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
ven_cat_t4 = tienda4.groupby('Categoría del Producto')['Categoría del Producto'].count().sort_values(ascending=False)
ven_cat_t4

new_var = plt.figure(figsize=(6,6))
# Lista de colores:
pie_colors = ['#711DB0', '#C21292', '#EF4040', '#FFA732', '#F5E965', '#FF9668', '#DB456F', '#A74FAF']
plt.pie(ven_cat_t4.values, labels=ven_cat_t4.index, autopct='%1.1f%%', shadow=True, startangle=140, colors=pie_colors)
plt.title('Ventas por categoría tienda 4')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


#### Calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
import matplotlib.pyplot as plt

#Obtener calificación por tienda:
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

print(f'La tienda 1 tiene una calificación de: {calificacion_tienda1:.2f}')
print(f'La tienda 2 tiene una calificación de: {calificacion_tienda2:.2f}')
print(f'La tienda 3 tiene una calificación de: {calificacion_tienda3:.2f}')
print(f'La tienda 4 tiene una calificación de: {calificacion_tienda4:.2f}')

evaluaciones = [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
tienas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Se colocan los colores dependiendo de la evaluación obtenida:
colores_evaluaciones = []
for evaluacion in evaluaciones:
  if evaluacion < 3.5:
    colores_evaluaciones.append('#DC0E0E')
  elif evaluacion < 4.0:
    colores_evaluaciones.append('#F87B1B')
  else:
    colores_evaluaciones.append('#F4F754')
plt.figure(figsize=(8,8))
barras = plt.bar(tiendas, evaluaciones, color= colores_evaluaciones)
plt.xticks(rotation=45, ha='right')
plt.title('Calificación promedio de las tiendas', fontsize=14, fontweight= 'bold')
plt.ylabel('Evaluaciones', fontsize=12)
plt.plot(tiendas, evaluaciones, marker='o', linestyle='--', color='red')

# Colocar la evaluación sobre las barras:
for barra in barras:
  height = barra.get_height()
  plt.text(barra.get_x() + barra.get_width()/2., height + 0.1, f'{height:.2f}', ha= 'center', va='bottom', fontsize=10, fontweight= 'bold')
```

#### La aprobación de las 4 tiendas se encuentra por arriba de la media, sin embargo las mejor punteadas son la 2 y 3. Quedando abajo la tienda 4 y 1.

# 5. Envío promedio por tienda

####calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
import matplotlib.pyplot as plt

costo_envio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_tienda4 = tienda4['Costo de envío'].mean()

print(f'El costo de envío promedio de la tienda 1 es: ${costo_envio_tienda1:.2f}')
print(f'El costo de envío promedio de la tienda 2 es: ${costo_envio_tienda2:.2f}')
print(f'El costo de envío promedio de la tienda 3 es: ${costo_envio_tienda3:.2f}')
print(f'El costo de envío promedio de la tienda 4 es: ${costo_envio_tienda4:.2f}')

plt.figure(figsize=(10,6))
costo_envios = [costo_envio_tienda1, costo_envio_tienda2, costo_envio_tienda3, costo_envio_tienda4]
tienas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['#F38181', '#FCE38A', '#EAFFD0', '#95E1D3']
barra = plt.barh(tiendas, costo_envios, color=colores)
plt.title('Costo de envio promedio por tienda', fontsize=14, fontweight= 'bold')
plt.xlabel('Costo', fontsize=12)
plt.ylabel('Tiendas', fontsize=12)

#Colocar valores en barras:

for bar in barra:
  cost_value = bar.get_width() # Para la barra horizontal, el ancho es el valor
  y_position = bar.get_y() + bar.get_height()/2 # Centro de la barra verticalmente
  x_position = bar.get_width() + 0.1 # Justo a la derecha de la barra

  plt.text(x_position, y_position, f'${cost_value:,.0f}', va='center', fontsize=8)
```
